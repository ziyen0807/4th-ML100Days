```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
arr = np.arange(-20,10.5,.5)

fx = lambda x:(x+5)**2
plt.plot(arr,fx(arr))


learning_rate = .1
iters = 1
pre_step_size = 1
while pre_step_size > precision and iters < max_iters:
    pre_cur = cur_x
    cur_x -= learning_rate * df(cur_x)
    pre_step_size = abs(pre_cur - cur_x)
    iters += 1
    print(pre_step_size)
    plt.plot(cur_x,fx(cur_x),c = 'r',marker = 'o')
    display.clear_output(wait = True)
    display.display(plt.gcf())
    
```
