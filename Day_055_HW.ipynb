- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X.shape)
print(iris.feature_names)
print(iris.target_names)
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

estimator = KMeans(n_clusters = len(iris.target_names))
estimator.fit(X)
predict_y = estimator.predict(X)
centers = estimator.cluster_centers_
plt.figure(figsize = (15,15))
for i1 in range(centers.shape[0]):
    for i2 in range(centers.shape[1]):
        plt.subplot(centers.shape[0],centers.shape[1],i1*3+i2+i1+1)
        plt.scatter(X[:,i1],X[:,i2],cmap = 'viridis',c = predict_y)
        plt.scatter(centers[:,i1],centers[:,i2],s = 200,c = 'r',alpha = .5)
        plt.xlabel(iris.feature_names[i1])
        plt.ylabel(iris.feature_names[i2])
        
plt.show()


```
